```python
import cv2
import numpy as np
import matplotlib.pyplot as plt 
import time
```

```python
def makeGaussian(size, fwhm = 0, center = None):
    fwhm = size / 4
    x = np.arange(0, size, 1, float)
    y = x[:,np.newaxis]

    if center is None:
        x0 = y0 = size // 2
    else:
        x0 = center[0]
        y0 = center[1]

    return np.exp(-4*np.log(2) * ((x-x0)**2 + (y-y0)**2) / fwhm**2)
```

```python
def calc_clahe(clip_limit, grid_size, img, img_bgr, z):
    start_time = time.time()
    
    clahe = cv2.createCLAHE(clipLimit=clip_limit, tileGridSize=(grid_size,grid_size))
    cl = clahe.apply(img)
    gaussian_blur = cv2.GaussianBlur(cl, (7, 7), 0)
    
    sub = gaussian_blur - img
    small_frame = cv2.resize(sub, (0, 0), fx=0.125, fy=0.125)
    height, width = small_frame.shape
    for i in range(height):
        for j in range(width):
            if (small_frame[i][j] > 100 ):
                small_frame[i][j] = small_frame[i][j] * 0
            continue 
        continue
    frame = cv2.resize(small_frame, (0, 0), fx = 8, fy = 8)
   
    ret,thresh = cv2.threshold(frame, 0, 1, cv2.THRESH_BINARY)
    img2, contours, hierarchy = cv2.findContours(thresh,cv2.RETR_TREE,cv2.CHAIN_APPROX_SIMPLE)
    croped_box = 0
    for i in range(int(len(contours))):
        (x,y,w,h) = cv2.boundingRect(contours[i])
        if (w > 100 and h > 100 and w < 700 and h < 700):
            box = x, y, x + w, y + h
            cropped_box = ROI(img_bgr, box, z)
#             print(box)
            cv2.rectangle(img_bgr, (x, y), (x + w, y + h), (0, 0, 255), 2)
                    
            example_img = cv2.imread('2.png', 3) # Поиск по примеру
            result = cv2.matchTemplate(cropped_box, example_img, cv2.TM_CCOEFF_NORMED)
            
            height_res, width_res = result.shape
                       
                
#             fh=int(height_res-1) / 2 - (height_res - 1) * 0.25
#             sh=int(height_res-1) / 2 + (height_res - 1) * 0.25
#             fw=int(width_res-1) / 2 - (width_res - 1) * 0.25
#             sw=int(width_res-1) / 2 + (width_res - 1) * 0.25
    
#             print((int((height_res-1) / 2 - (height_res - 1) * 0.25)))
#             print((int((height_res-1) / 2 + (height_res - 1) * 0.25)))
#             print( int(((width_res-1) / 2 - (width_res - 1) * 0.25)))
#             print( int(((width_res-1) / 2 + (width_res - 1) * 0.25)))
    
            zzz = makeGaussian(width_res)
    
            zzz1 = zzz[0:height_res-1][0:width_res-1]
            zzz2 = result[0:height_res-1][0:width_res-1]     
            zzz3 = zzz1 * zzz2
            
            cv2.imshow('img_bgr',img_bgr)
            cv2.imshow('result',zzz3)
            
            height3,width3 = zzz3.shape
            summ = 0
            for i in range(height3):
                for j in range(width3):
                    summ += zzz3[i][j]
                    continue
                continue
#             cv2.imshow('gaussian', zzz1)
            cv2.imshow('search_with_example', zzz2)
#             cv2.waitKey(200)
            print('summ, ', summ)
            if summ > 20:
                print('блик\n')
            elif summ < 20:
                print('нет блика\n')
            
        continue
        end_time = time.time()
        print(end_time - start_time)
    return (img_bgr, thresh)
```

```python
def ROI(frame, box, z):
    CroppedBox = frame[box[1] : box[3], box[0] : box[2]]
    cv2.imwrite(str("C://Users//am.miliutin//Desktop//Crop//file_" + str(z) + ".png"), CroppedBox)
    return CroppedBox    
```

```python
cap = cv2.VideoCapture("2.cva")
z = 0
while(True):
    ret, img_bgr = cap.read()
    img = cv2.cvtColor(img_bgr, cv2.COLOR_BGR2GRAY)
    frame, thresh = calc_clahe(10, 100, img, img_bgr, z)    
#     cv2.imshow('frame', frame)
    z += 1
   
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break
        
cap.release()
cv2.destroyAllWindows()
```
